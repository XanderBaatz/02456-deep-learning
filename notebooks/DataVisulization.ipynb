```python
import numpy as np
import tarfile
import pickle
import os
import matplotlib.pyplot as plt
import seaborn as sns

import sys
sys.path.append("../")
from dataloader import load_cifar10
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load CIFAR-10
X_train, y_train, X_val, y_val, X_test, y_test = load_cifar10("../data/cifar-10-python.tar.gz")

# Select one random image per class from the training set
selected_images = []
selected_labels = []

for label in range(10):
    idx = np.random.choice(np.where(y_train == label)[0])
    selected_images.append(X_train[idx])
    selected_labels.append(label)

# Plot images side by side
fig, axes = plt.subplots(1, 10, figsize=(20, 5))

for i in range(10):
    img = selected_images[i].transpose(1, 2, 0)  # (32, 32, 3)
    axes[i].imshow(img)
    axes[i].set_title(f"Label {selected_labels[i]}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
